```python
import os
import pandas as pd
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
path_to_dataset = 'Salary_dataset.csv'
if not os.path.isfile(path_to_dataset):
    print('not a file')
```

```python
data = pd.read_csv(path_to_dataset)
data.head()
```

```python
data = data.drop(columns=['Unnamed: 0'])
```

```python
data.head()
```

```python
plt.scatter(data['YearsExperience'], data['Salary'], color='blue')
plt.xlabel('Ani Experienta')
plt.ylabel('Salariu')
plt.title('Salariu vs Experienta')
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = data[['YearsExperience']]
y = data['Salary']

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

plt.scatter(X, y, color='blue', label='Date reale')
plt.plot(X, y_pred, color='red', label='Regresie liniara')
plt.xlabel('Ani Experienta')
plt.ylabel('Salariu')
plt.title('Regresie Liniara: Salariu vs Experienta')
plt.legend()
plt.grid(True)
plt.show()
```

### Calculul regresiei doar cu numpy
- Formula
    - y = mX + b
- Calculam coeficientii
    - m -> panta
    - b -> valoarea lui y pentru care linia intersecteaza axa Y

```python
data = pd.read_csv(path_to_dataset)
data = data.drop(columns=['Unnamed: 0'])

X = data['YearsExperience'].values
y = data['Salary'].values
```

```python
X
```

```python
x_mean = np.mean(X)
x_mean
```

```python
y
```

```python
y_mean = np.mean(y)
y_mean
```

### Calculam panta m

```python
a = np.sum((X - x_mean) * (y - y_mean))
b = np.sum((X - x_mean)**2)
m = a / b
print(f'panta este: {m}')
```

### Calculam interceptia b

```python
b = y_mean - m * x_mean
```

```python

print(f"Coeficient (m): {m}")
print(f"Interceptie (b): {b}")
```

```python
y_pred = m * X + b
```

```python
plt.scatter(X, y, color='blue', label='Date reale')
plt.plot(X, y_pred, color='red', label='Regresie')
plt.xlabel('Ani Experienta')
plt.ylabel('Salariu')
plt.title('Regresie Liniara')
plt.legend()
plt.grid(True)
plt.show()
```
